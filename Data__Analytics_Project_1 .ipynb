```python
#Importing all the Libraries needed
import pandas as pd
import numpy as np
```

```python
# Read the Excel sheet
df = pd.read_excel("/content/Dataset for Data Analytics(project1).xlsx")
```

```python
#Printing the Shape and rows and colums
print(f"The Shape of the Dataset : {df.shape}\n")
print(f"No.of Rows in the Dataset : {df.shape[0]}\n")
print(f"No.of Columns of the Dataset : {df.shape[1]}\n")
```

```python
#Printing the info of DataSet
df.info()
```

```python
print('Unique values count for each column:')
for column in df.columns:
  print(f'{column}: {df[column].nunique()}')
```

```python
#Finding the NULL Values Existance
df.isna().sum()
```

```python
#Filling Null Values with i.e The column CouponCode With the text NA
df['CouponCode'] = df['CouponCode'].fillna('Not Available(NA)')
df['CouponCode'].head(10)
```

```python
#Checking Again for the Null Values
df.isna().sum()
```

```python
# Checking for duplicate OrderIDs
duplicate_order_ids = df[df['OrderID'].duplicated()]
print(f"Number of duplicate OrderIDs: {len(duplicate_order_ids)}")

# Checking for duplicate CustomerIDs
duplicate_customer_ids = df[df['CustomerID'].duplicated()]
print(f"Number of duplicate CustomerIDs: {len(duplicate_customer_ids)}")
```

```python
#Printing if , is there any Duplicated Values
print(df.duplicated().sum())
```

```python
# Corrected date handling: Convert to datetime and keep as datetime objects
df["Date"] = pd.to_datetime(df["Date"])
print("Date column converted to datetime objects.")
display(df['Date'].head())
```

```python
# Remove Extra Spaces
text_columns = [
    "Product",
    "ShippingAddress",
    "PaymentMethod",
    "OrderStatus",
    "TrackingNumber",
    "CouponCode",
    "ReferralSource"
]

for col in text_columns:
    df[col] = df[col].astype(str).str.strip()

df.head(10)
```

```python
#Checking Whether there are any extra spaced elements
for col in text_columns:
    count = (df[col].astype(str) != df[col].astype(str).str.strip()).sum()
    print(col, ":", count)
```

```python
float_cols = df.select_dtypes(include = "float64").columns
df[float_cols] = df[float_cols].round(1)
display(df[float_cols].head())
```

```python
#Final Check
df.head()
```

```python
#Exporting into New File
df.to_excel("Cleaned_Dataset.xlsx", index=False)
```
